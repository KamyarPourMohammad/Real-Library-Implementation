```python
class Library:
    def __init__(self, name, established_year):
        self.name = name
        self.established_year = established_year
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)
        else:
            print("Book not found")

    def list_books(self):
        return self.books

    def number_of_books(self):
        return len(self.books)
```

```python
class Admin:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.library = None

    def create_library(self, name, year):
        self.library = Library(name, year)
```

```python
def run_app():
    name = input("Admin name: ")
    age = int(input("Admin age: "))
    admin = Admin(name, age)

    lib_name = input("Library name: ")
    year = int(input("Established year: "))
    admin.create_library(lib_name, year)

    while True:
        print("\n1 Add book")
        print("2 Remove book")
        print("3 Show books")
        print("4 Exit")

        choice = input("Choice: ")

        if choice == "1":
            admin.library.add_book(input("Book: "))
        elif choice == "2":
            admin.library.remove_book(input("Book: "))
        elif choice == "3":
            print(admin.library.list_books())
        elif choice == "4":
            break
```

```python
run_app()
```
